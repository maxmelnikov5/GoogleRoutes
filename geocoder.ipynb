```python
# functions
def get_geocode(adr:str, api_key:str ):

    gmaps = googlemaps.Client(key=api_key)
    geocode_result = gmaps.geocode(adr, language="ru")

    df = pd.json_normalize(geocode_result).assign(
    locality=lambda x: x['address_components'].apply(lambda comps: next((comp['long_name'] for comp in comps if 'locality' in comp['types']), None)),
    administrative_area_level_2=lambda x: x['address_components'].apply(lambda comps: next((comp['long_name'] for comp in comps if 'administrative_area_level_2' in comp['types']), None)),
    administrative_area_level_1=lambda x: x['address_components'].apply(lambda comps: next((comp['long_name'] for comp in comps if 'administrative_area_level_1' in comp['types']), None)),
    route=lambda x: x['address_components'].apply(lambda comps: next((comp['long_name'] for comp in comps if 'route' in comp['types']), None)),
    street_number=lambda x: x['address_components'].apply(lambda comps: next((comp['long_name'] for comp in comps if 'street_number' in comp['types']), None))
        )[[
        'geometry.location.lat', 'geometry.location.lng', 'formatted_address', 'locality', 'administrative_area_level_2', 'administrative_area_level_1', 'route', 'street_number'
        ]]
    
    return df.iloc[0].to_dict()
```

```python
adr_list = pd.read_csv('source\\tShops (1).csv')
# adr_list['adr_tocheck'] = adr_list['Город']+" "+adr_list['Адрес']
adr_list['adr_tocheck'] = adr_list['Город'].fillna('').astype(str) + " " + adr_list['Адрес'].fillna('').astype(str)
adr_list = adr_list[adr_list['Адрес'] != '#Н/Д']

adr_list
```

```python
# Apply get_geocode to each address and create a DataFrame from the result dictionaries
geocode_results = adr_list['adr_tocheck'].apply(lambda adr: get_geocode(adr, APIkey))
geocode_df = pd.DataFrame(geocode_results.tolist(), index=adr_list.index)

# Concatenate the new geocode columns to the original DataFrame
adr_list2 = pd.concat([adr_list, geocode_df], axis=1)
adr_list2
```
